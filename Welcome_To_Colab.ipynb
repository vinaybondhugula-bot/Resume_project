```
data = []
for file in os.listdir():
  if file.lower().endswith(".pdf"):
    print("Reading",file)

    with open(file, 'rb') as f:
      reader = PyPDF2.PdfReader(f)
      text = ""
      for page in reader.pages:
        if page.extract_text():
          text += page.extract_text()
    data.append({
        "file_name":file,
        "resume_text":text
    })
```

```
df = pd.DataFrame(data)
df.head()
```

```
skills_list = [
    "python", "sql", "java", "javascript", "react", "angular",
    "node", "django", "flask", "sap", "selenium", "cypress",
    "manual testing", "automation testing", "api testing",
    "aws", "azure", "docker", "kubernetes", ".net","jira","soql"
]
```

```
def extract_skills(text):
  found_skills = set()
  text = text.lower()
  for skill in skills_list:
      if skill in text:
          found_skills.add(skill)
  return ", ".join(found_skills)
```

```
df["skills"] = df["resume_text"].apply(extract_skills)
df.head()
```

```
def extract_education(text):
    text = text.lower()

    if any(word in text for word in [
        "mtech", "m.tech", "mba", "master"
    ]):
        return "Masters"
    elif any(word in text for word in [
        "btech", "b.tech", "be", "b.e",
        "bsc", "b.sc", "bachelor","bachelor's"
    ]):
        return "Bachelors"

    else:
        return "Other"
```

```
df["education"] = df["resume_text"].apply(extract_education)
df.head()
```

```
df["skills_count"] = [len(x.split(","))if x else 0 for x in df["skills"]]
df.head()
```

```
def extract_certification(text):
  text = text.lower()
  keywords = ["certification","certified","certifications"]
  return 1 if any(keyword in text for keyword in keywords) else 0

df["certification"] = df["resume_text"].apply(extract_certification)
df.head()
```

```
df = df[df["resume_text"].str.strip() != ""]
```

```
df.head()
```

```
def extract_projects(text):
  return text.lower().count("project")
df["project"] = df["resume_text"].apply(extract_projects)
```

```
df.head(3)
```

```
import re
def extract_experience(text):
    if not isinstance(text, str):
        return 0.0
    text = text.lower()
    text = re.sub(r'(\d)\s*\.\s*(\d)',r'\1.\2', text)
    matches = re.findall(r'(\d+(?:\.\d+)?)\+?\s*(years|year|yrs|yr)', text)

    for m in matches:
      val = float(m[0])
      if 0<val<20:
        return val
    return 0.0
```

```
df["experience"] = df["resume_text"].apply(extract_experience)
df.head(3)
```

```
df = df.rename(columns={"project":"project_count"})
df.head(3)
```

```
edu_map = {
    "Masters":2,
    "Bachelors":1,
    "Others":0
}
df["education_rank"] = df["education"].map(edu_map).fillna(0)
```

```
df.head(2)
```

```
y = df["experience"]
X= df.drop(columns=["file_name","resume_text","skills","education","experience"])
```

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```
model.score(X_test,y_test)
```

```
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split':[2,5,10]
}

rf = RandomForestRegressor()
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, scoring='r2')
grid_search.fit(X_train, y_train)
```

```
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split':[2,5,10]
}

rf = RandomForestRegressor()
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=2, scoring='r2')
grid_search.fit(X_train, y_train)
```
